```python
#21
def smoothedY(y, n):
    y_s = []
    
    y_s.append(y[0])
    y_s.append((18*y[0] + 26*y[1] + 24*y[2] + 12*y[3] - 10*y[4])/70)
    for i in range(2, n-1):
        y_s.append((-3*y[i-2] + 12*y[i-1] + 17*y[i] + 12*y[i+1] - 3*y[i+2])/35)
    y_s.append((-10*y[n-4] + 12*y[n-3] + 24*y[n-2] + 26*y[n-1] + 18*y[n])/70)
    y_s.append(y[n])
    
    return y_s
h = 2 
x = np.arange(-4, 6, h)
y = x**2
plt.plot(x, y, c='red')
h = 2 
x_new = np.arange(-4, 4.1, 0.1)
y_new = x_new**2
plt.grid()
plt.plot(x_new, smoothedY(y_new, len(y_new)-1), c='black', linestyle='dashed')
```

```python
#23
def smoothedY(y, n):
    y_s = []
    
    y_s.append(y[0])
    y_s.append((y[0] + 1858*y[1] + 720*y[2] - 718*y[3] + 239*y[4])/2100)
    for i in range(2, n-1):
        y_s.append((-3*y[i-2] + 12*y[i-1] + 17*y[i] + 12*y[i+1] - 3*y[i+2])/35)
    y_s.append((239*y[n-4] - 718*y[n-3] + 720*y[n-2] + 1858*y[n-1] + y[n])/2100)
    y_s.append(y[n])
    
    return y_s

h = 2 
x = np.arange(-4, 6, h)
y = x**3
plt.plot(x, y, c='red')

h = 2 
x_new = np.arange(-4, 4.1, 0.1)
y_new = x_new**3 
plt.grid()
plt.plot(x_new, smoothedY(y_new, len(y_new)-1), c='black', linestyle='dashed')
```

```python
#22
def findCoef(x, y, h):
    s = []
    s.append(5)
    s.append(x[0] + x[1] + x[2] + x[3] + x[4]) 
    s.append(x[0]**2 + x[1]**2 + x[2]**2 + x[3]**2 + x[4]**2)
    s.append(x[0]**3 + x[1]**3 + x[2]**3 + x[3]**3 + x[4]**3)
    s.append(x[0]**4 + x[1]**4 + x[2]**4 + x[3]**4 + x[4]**4)
   
    t = []
    t.append(y[0] + y[1] + y[2] + y[3] + y[4])
    t.append(y[0]*x[0] + y[1]*x[1] + y[2]*x[2] + y[3]*x[3] + y[4]*x[4])
    t.append(y[0]*x[0]**2 + y[1]*x[1]**2 + y[2]*x[2]**2 + y[3]*x[3]**2 + y[4]*x[4]**2)
    
    a = [0, 0 , 0]
    
    a[2] = t[2] - t[0]*s[2]/5 - (t[1] - t[0]*s[1]/5)*(s[3] - s[1]*s[2]/5)/(s[2] - s[1]**2/5)
    a[2] /= (s[4] - s[2]**2/5 - (s[3]-s[1]*s[2]/5)**2/(s[2] - s[1]**2/5))
    
    a[1] = t[1] - t[0]*s[1]/5 - (s[3] - s[2]*s[1]/5)*a[2]
    a[1] /= s[2] - s[1]**2/5
    
    a[0] = (t[0] - s[2]*a[2] - s[1]*a[1])/5
    return a
def smoothedY(x, n, a, h):
    y_s = []
    for i in range(0, n+1):
       y_s.append(a[0] + a[1]*x[i] + a[2]*x[i]**2)
    return y_s
h = 2 
x = np.arange(-4, 6, h)
y = x**2 - x + 6
plt.plot(x, y, c='red')
a = findCoef(x, y, h)
h = 2 
x_new = np.arange(-4, 4.1, 0.1)
plt.grid()
print(findCoef(x, y, h))
plt.plot(x_new, smoothedY(x_new, len(x_new)-1, a, h), c='black', linestyle='dashed')
```
